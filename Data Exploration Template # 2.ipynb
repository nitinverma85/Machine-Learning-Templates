# ML Basics

#### Univariate, Bivariate and Multivariate
When we are classifying a label/class/output variable based on data of 1 variable, 2 or multiple features. 

### Classifying iris dataset using Univariate, Bivariate and Multivariate analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

df = pd.read_csv("iris.csv")
df.head()
```

#### Univariate analysis

```python
plt.scatter(x=df[df['species']=='Iris-setosa']['sepal_width'],y=np.zeros(df[df['species']=='Iris-setosa'].count()[0]))
plt.scatter(x=df[df['species']=='Iris-versicolor']['sepal_width'],y=np.zeros(df[df['species']=='Iris-versicolor'].count()[0]))
plt.scatter(x=df[df['species']=='Iris-virginica']['sepal_width'],y=np.zeros(df[df['species']=='Iris-virginica'].count()[0]))
```

#### Bivariate analysis

```python
sns.pairplot(data=df,hue='species',size=4)
```

```python
sns.lmplot(data=df,x='petal_width',y='sepal_width',hue='species',fit_reg=False)
```
