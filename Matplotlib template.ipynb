# Matplotlib

#### Importing and basics

```python
import matplotlib.pyplot as plt
import numpy as np #No need to write plt.show if you write this line
%matplotlib inline 

x=np.arange(10)
y=x*x
```

#### Scatter charts

```python
plt.scatter(x,y,c='b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('This is my chart')
plt.savefig('test_delete.png')
```

#### Normal line chart

```python
plt.plot(x,y,'ro--')
plt.subplot(2,2,1)
plt.plot(x,x*x,'r')
plt.subplot(2,2,2)
plt.plot(x,x*x,'b')
plt.subplot(2,2,3)
plt.plot(x,x*x,'g')
plt.subplot(2,2,4)
plt.plot(x,x*x,'y')
```

#### Bar chart

```python
x1=[1,2,3]
y1=[4,5,6]
x2=[4,5,6]
y2=[7,8,9]
plt.bar(x1,y1,color='g')
plt.bar(x2,y2)
```

#### Histograms

```python
x=[1,1,2,3,4,5,6,2,3,4,4]
plt.hist(x)
```

#### Box plots

```python
data=[np.random.normal(1000,std,100) for std in range (1,4)] # np.random.normal (mean, std dev, number of samples)
plt.boxplot(data) # Plots 0,25,50,75 and 100 percentile
```

#### Pie Charts

```python
labels=['A','B','C','D']
values=[10,20,30,40]
plt.pie(values,labels=labels)
```
