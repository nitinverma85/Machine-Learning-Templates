# Seaborn basics
#### Importing new dataset

```python
import seaborn as sns
df=sns.load_dataset('tips')
df.head() #Tip is the dependent feature
```

## Plots for numeric variables

#### Correlation

```python
df.corr()
#df.dtypes
sns.heatmap(df.corr())
```

#### Joint plot

```python
sns.jointplot(x='tip',y='total_bill', data=df,kind='reg')
```

#### Pair plot

```python
sns.pairplot(df,hue='time')
```

#### Histograms (Also called as dist plots)

```python
sns.distplot(df['tip'],kde=False)
```

## Categorical Variables

#### Countplot

```python
sns.countplot('sex',data=df,hue='day') # Behaves as value_counts
```

#### Bar plot

```python
sns.barplot(x='total_bill',y='smoker',data=df) # Gives avg
```

#### Box plot

```python
#sns.boxplot(x='sex',y='total_bill',data=df)
sns.boxplot(data=df)
```

#### Violin plot

```python
sns.violinplot(x='sex',y='total_bill',data=df)
```
