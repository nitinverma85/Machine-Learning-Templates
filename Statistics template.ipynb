#### Distribution (Probability density function or pdf)
1. Convert histogram into a curve. If the curve is centered around mean and contains 68% of the values within 1 stdev, 98% within 2, it is called Gaussian. 

2. Area under the chart is the percentile. Or if you convert it into a cumulative chart (Cumulative density function or cdf), then you can simply check the y value to find the percentile for a particular value

#### Gaussian Distribution

Objective of ML is to have all features on the same scale to increase accuracy of the model. If all variables are on the same scale , then accuracy of the models is increased because the cost function is spread over a much smaller area and it is easier to find the minima. 

There are 2 methods to bring all variables on the same scale

##### Standardisation (Standard Scaler in Python)
We subtract mean from all values and divide by standard deviation (this is also called as Zscore). Final values are not always between -1 and 1. 

1. Distribution: Plot histogram for a variable. The curve formed is called as a distribution. 
2. Standard deviation : How far are your values from the mean. So, for example, a random distribution between 1 to 100 has a mean of 50 and a stdev of 29.  
3. Gaussian distribution: Gaussian or Normal distribution is a bell curve that has 68% of the values between mean+stdev and mean-stdev, 95% of the values between mean+2*stdev and mean-2*stdev. For example, a random distribution between 1 to 100 is not a Gaussian or Normal distribution since numbers are spread evenly, not centered around mean. Heights of peole in the world will form a distribution very close to Gaussian/Normal distribution. 
3. Standard Normal Distribution: Gaussian distribution whose mean is 0 and stdev is 1.We subtract the mean and divide by stdev to convert all values to Standard Normal Distribution in case they are already in Gaussian distribution.   
4. Log normal: When log of a variable forms a Gaussian Distribution. The bell curve of this distribution is right skewed (doesn't fall to 0 within 2-3 stdev but takes more stdev). Example could be income of people, since on the left of mean, it will fall to 0, but on the right, it will continue until Ambanis of the world..:)

##### Normalisation (min max scaler in python)

We use a min max scaler here. (x-x(min))/x(max)-x(min). Final values are always between 0 and 1.  

#### Z-Score 
Each value in a variable can be converted to a z-score (how many stdev away from the mean). For example, in a mean of 50 and stdev of 5, 47.5 will have a Zscore of -.5. We have a Zscore table which helps us to find the % of values below a particular Zscore of .308. So, around 70% of the values will be above 47.5 in case of Gaussian distribution.  

#### Covariance
1. Calculated between 2 variables
2. Calculated as (x-mean(x))*(y-mean(y)) for all values
3. If positive correlation (x increasing means y increasing or vica versa), covariance is +ve, else -ve

#### Correlation (Pearson)
Correlation is defined as Covariance divided by variance of 1st var and variance of 2nd var (to bring it between 1 and -1)

#### Correlation (Spearman)
This is better if the relationship between x and y is non linear. If, for example, x=y*y, Pearson correlation will be .9 while Spearman correlation will be 1

#### Outliers
1. Zscore method: If the value in a variable is not falling within 3 standard deviations towards right or left, it can be considered as an outlier. Zscore is the number of standard deviation a variable is away from the mean. 
2. Interquartile method: If the value in a variable is not falling within 25% quartile-()*1.5 and 75% quartile + ()*1.5, then it can be considered as an outlier. ()=75%Quartile value-25% quartile value. () is also called as interquartile range. 

#### Bias and Variance
Bias of a model means how well it is able to match the training data. Variance is how it is able to match test data. 

If undefitting, it will have high bias and high variance. 
If overfitting, it will have low bias and high variance. 

#### Multicollinearity
If independent features have high correlation among themselves

#### R-squared and Adjusted R-squared 
1. It's a method of checking accuracy for regression models. 
2. Fundamentally, it is 1-(cost function normalised)
1-((y(actual)-y(pred))**2)/((y(actual)-y(mean))**2
3. Value closer to 1 is better
4. Adjusted R squared: When we keep on adding independent variables, R-squared value can keep on increasing because of overfitting. TO penalise, new variables, we have adjusted R-squared which doesn't increase as the number of variables increase. 

#### Cross validation
To bring randomness to test and train data, we pick test and train data multiple times. This is called as cross validation. k-cross validation means the model will pick test and train data randomly for k times. Average of all models will be used. 

#### Hypothesis testing, type 1 error, type 2 error
Null hypothesis, Alternate hypothesis. Example, 

Null hypothesis: The consumer will not leave
Alternate hypothesis: The consumer will leave. 

Type 1 error: The consumer leaves
Type 1 errors are fatal.

Type 2 error: The consumer doesn't leave
Type 2 errors are still ok
